```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fichero1='netflix_titles.csv'

df = pd.read_csv(fichero1,index_col='show_id')

df.head()
#df.info()
```

```python
df['type'].value_counts().plot(kind='pie',autopct='%1.0f%%', title='Distribution of Show/Movie')
plt.show()
```

```python
df['release_year'].plot(kind='bar')
plt.show()
```

```python
#grafico de los 10 años con más producciones de películas


tries = pd.DataFrame(df['release_year'].value_counts().nlargest(10).sort_index())
tries = tries.reset_index()
tries['index'] = tries['index'].apply(str)
tries.plot(x='index', kind='line', title='Cant. de producciones por año')

plt.show()
```

```python
def fill(frame,col):
    frame[col].fillna('ND',inplace=True)
    return frame.head()

fill(df,'director')
fill(df,'cast')
```

```python
df.head()
```

```python
tries = pd.DataFrame(df['director'].dropna().value_counts().nlargest(5).sort_index())
tries = tries.reset_index()
tries['index'] = tries['index'].apply(str)
tries['director'] = tries['director'].apply(int)
tries.plot(x='index', y='director',kind='bar', title='Cant. de producciones por año')
```
